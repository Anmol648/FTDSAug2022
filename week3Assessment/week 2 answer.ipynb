```python
#Problem 2b
def makeCategory(x):
    if x=="Below $50":
        return 'Cheap'
    elif x=="$51-100":
        return 'Not Cheap'
    elif x=="$101-200":
        return 'Expensive'
    else:
        return 'Very Expensive'
    
df['price_category'] = df['price_range'].apply(makeCategory)
df.head(3)
```
